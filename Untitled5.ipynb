```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from xgboost import XGBRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score

# --- Step 1: Generate mock NVIDIA stock data ---
np.random.seed(42)
days = pd.date_range(end=pd.Timestamp.today(), periods=60, freq='B')
start_price = 650
price_changes = np.random.normal(0, 5, len(days))
prices = start_price + np.cumsum(price_changes)

# Put into a DataFrame
data = pd.DataFrame({
    'Date': days,
    'Price': [f"{p:,.2f}" for p in prices]  # formatted with commas
})

# Save to CSV to simulate real data
csv_file = "NVIDIA Stock Price History.csv"
data.to_csv(csv_file, index=False)
print("✅ Data saved.")

# --- Step 2: Load and clean the data ---
try:
    df = pd.read_csv(csv_file)
    print("CSV loaded.")
except:
    print("File couldn’t be loaded.")
    exit()

df['Date'] = pd.to_datetime(df['Date'])

# 🔧 Convert 'Price' to numeric by stripping commas
df['Close'] = df['Price'].astype(str).str.replace(',', '').astype(float)

# Sort by date
df = df.sort_values('Date')
df = df.dropna(subset=['Close'])
print("✅ Cleaned data shape:", df.shape)

# --- Step 3: Feature Engineering ---
# Create lag features (previous 3 days)
for i in range(1, 4):
    df[f'Lag_{i}'] = df['Close'].shift(i)

# Add simple rolling stats
df['Rolling_Mean'] = df['Close'].rolling(window=3).mean().shift(1)
df['Rolling_Std'] = df['Close'].rolling(window=3).std().shift(1)

# Drop rows with missing values due to lagging
df.dropna(inplace=True)
print("Final dataset shape:", df.shape)

# --- Step 4: Prepare data for modeling ---
features = [f'Lag_{i}' for i in range(1, 4)] + ['Rolling_Mean', 'Rolling_Std']
X = df[features]
y = df['Close']
dates = df['Date']

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# --- Step 5: Train-test split ---
split_point = int(len(df) * 0.8)
X_train, X_test = X_scaled[:split_point], X_scaled[split_point:]
y_train, y_test = y[:split_point], y[split_point:]
dates_test = dates.iloc[split_point:]

# --- Step 6: Train the model ---
model = XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=0)
model.fit(X_train, y_train)

# --- Step 7: Predictions and Evaluation ---
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"MSE: {mse:.2f}")
print(f"R² Score: {r2:.2f}")

# --- Step 8: Plot results ---
plt.figure(figsize=(12, 6))
plt.plot(dates_test, y_test.values, label="Actual", color='blue', linewidth=2)
plt.plot(dates_test, y_pred, label="Predicted", color='orange', linewidth=2)
plt.title("NVIDIA Stock Price - Actual vs Predicted")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
